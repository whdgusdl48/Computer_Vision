```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('2015_2019data.csv')
data
```

```python
data.info()
```

```python
means = []
year = np.array(list(set(data['year'])))
month = np.array(list(set(data['month'])))
data2 = np.array(data)
print(year,month)
for i in range(len(year)):
  for j in range(len(month)):
    t_y = year[i]
    t_m = month[j]
    count = 0
    sum = 0
    for k in range(len(data2)):
      if data2[k][1] == t_y and data2[k][2] == t_m:
        sum += (data2[k][5])
        count += 1
      else:
        continue
    if count == 0:
      means.append(0)
    else:
      means.append(sum / count)
    count = 0
print(means)
s = []

for i in range(5):
  t = means[i * 12 : (i+1)*12]
  s.append(t)

for i in range(5):
  plt.figure()
  plt.plot(month,s[i])
  st = str(year[i])
  plt.title(st)
  plt.savefig(st + 'png')
```
